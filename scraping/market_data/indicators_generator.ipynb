```python
import pandas as pd 
import matplotlib.pyplot as plt
import ta
import numpy
import os
```

```python

df = pd.read_excel(r"C:\Users\nextrow\OneDrive - NextRow Inc\Desktop\AGENTS\Stock_prediction\Stock_prediction\stock-ai-system\data\raw\market_data\indian_stocks.xlsx", sheet_name=None)
```

```python
def sort_data(stock_df):
    stock_df = stock_df.sort_values(["date"])
  
    return stock_df
```

```python
def add_indicators(stock_df):
    stock_df["sma_7"]=stock_df["close"].rolling(window=7).mean()
    stock_df["sma_20"]=stock_df["close"].rolling(window=20).mean()
    stock_df["sma_50"]=stock_df["close"].rolling(window=50).mean()
    stock_df["ema_9"]=stock_df["close"].ewm(span=9,adjust=False).mean()
    stock_df["ema_21"]=stock_df["close"].ewm(span=21,adjust=False).mean()
    stock_df["ema_50"]=stock_df["close"].ewm(span=50,adjust=False).mean()
    stock_df['macd']=ta.trend.MACD(stock_df['close']).macd()
    stock_df['rsi']=ta.momentum.RSIIndicator(stock_df['close']).rsi()
    stock_df['bollinger_bands_upper'] = ta.volatility.BollingerBands(stock_df["close"]).bollinger_hband()
    stock_df['bollinger_bands_lower'] = ta.volatility.BollingerBands(stock_df["close"]).bollinger_lband()

    return stock_df
```

```python

stock_df=[]
for symbol,data in df.items():
    data_sorted = sort_data(data)
    stock_df.append(add_indicators(data_sorted))
stock_df=pd.concat(stock_df,ignore_index=True)
```

```python
print(stock_df.groupby("stock")["close"].mean().describe())
print("\nTop 10 highest priced stocks:")
print(stock_df.groupby("stock")["close"].mean().sort_values(ascending=False).head(10))
print("\nBottom 10 lowest priced stocks:")
print(stock_df.groupby("stock")["close"].mean().sort_values().head(10))
```

```python
stock_df["date"] = pd.to_datetime(stock_df["date"])
stock_df["time_idx"]=stock_df.groupby("stock").cumcount()
stock_df.head()
```

```python
train_list, val_list, test_list = [], [], []

for stock, group in stock_df.groupby("stock"):
    group = group.sort_values("date")
    
    train_list.append(group[group["date"] <= "2023-06-30"])
    val_list.append(group[(group["date"] > "2023-06-30") & (group["date"] <= "2024-06-30")])
    test_list.append(group[group["date"] > "2024-06-30"])

train      = pd.concat(train_list).reset_index(drop=True)
validation = pd.concat(val_list).reset_index(drop=True)
test       = pd.concat(test_list).reset_index(drop=True)

print("Train:", train["date"].min(), "→", train["date"].max())
print("Val:  ", validation["date"].min(), "→", validation["date"].max())
print("Test: ", test["date"].min(), "→", test["date"].max())
print("Train stocks:", train["stock"].nunique())
print("Val stocks:  ", validation["stock"].nunique())
print("Test stocks: ", test["stock"].nunique())
print("Train rows:", len(train))
print("Val rows:  ", len(validation))
print("Test rows: ", len(test))
```

```python

train_stocks = set(train["stock"].unique())


validation = validation[validation["stock"].isin(train_stocks)].reset_index(drop=True)
test = test[test["stock"].isin(train_stocks)].reset_index(drop=True)


all_stocks = set(stock_df["stock"].unique())
removed = all_stocks - train_stocks
print("Removed stocks:", removed)
print("Remaining stocks:", len(train_stocks))
```

```python
from pytorch_forecasting import TimeSeriesDataSet
from pytorch_forecasting.data import GroupNormalizer
import pickle
training = TimeSeriesDataSet(
    train,
    time_idx="time_idx",
    target="close",
    group_ids=["stock"],
    max_encoder_length=60,
    max_prediction_length=7,
    static_categoricals=["stock"],
    time_varying_unknown_reals=[
        'open', 'high', 'low', 'close', 'volume',
        'sma_7', 'sma_20', 'sma_50',
        'ema_9', 'ema_21', 'ema_50',
        'macd', 'rsi',
        'bollinger_bands_upper', 'bollinger_bands_lower'
    ],
    target_normalizer=GroupNormalizer(groups=["stock"], transformation="softplus"),
    add_relative_time_idx=True,
    add_target_scales=True,
    add_encoder_length=True,
)

with open(r"C:\Users\nextrow\OneDrive - NextRow Inc\Desktop\AGENTS\Stock_prediction\Stock_prediction\stock-ai-system\ml_models\technical_prediction\training_dataset.pkl", "wb") as f:
    pickle.dump(training, f)
size = os.path.getsize(r"C:\Users\nextrow\OneDrive - NextRow Inc\Desktop\AGENTS\Stock_prediction\Stock_prediction\stock-ai-system\ml_models\technical_prediction\training_dataset.pkl")/1024/1024
print(f"✅ Saved: {size:.1f} MB")
```

```python
#datasets
testing = TimeSeriesDataSet.from_dataset(training,test,predict=True)
validating = TimeSeriesDataSet.from_dataset(training,validation,predict=True)
```

```python
#data loaders

training_dataloader   = training.to_dataloader(train=True,  batch_size=256, num_workers=2)
validation_dataloader = validating.to_dataloader(train=False, batch_size=256, num_workers=2)
testing_dataloader    = testing.to_dataloader(train=False,  batch_size=256, num_workers=2)
```

```python
from pytorch_forecasting import TemporalFusionTransformer
from pytorch_forecasting.metrics.quantile import QuantileLoss
tft = TemporalFusionTransformer.from_dataset(
    training,
    hidden_size=64,
    attention_head_size=4,
    dropout=0.1,
    hidden_continuous_size=32,
    learning_rate=0.003,          
    loss=QuantileLoss(),
    log_interval=10,
    reduce_on_plateau_patience=3   
)
```

```python
from lightning.pytorch.callbacks import Callback, ModelCheckpoint

class EpochLogger(Callback):
    def on_train_batch_end(self, trainer, pl_module, outputs, batch, batch_idx):
        if batch_idx % 100 == 0:
            loss = outputs['loss'].item()
            print(f"Epoch {trainer.current_epoch+1} | Batch {batch_idx} | train_loss: {loss:.4f}")

    def on_train_epoch_end(self, trainer, pl_module):
        train_loss = trainer.callback_metrics.get('train_loss_epoch', 0)
        val_loss = trainer.callback_metrics.get('val_loss', 0)
        print(f"\n Epoch {trainer.current_epoch+1}/30 | train_loss: {train_loss:.4f} | val_loss: {val_loss:.4f}\n")

checkpoint_callback = ModelCheckpoint(
    monitor="val_loss",
    dirpath="/kaggle/working/tft_checkpoints/",  
    filename="tft-best",
    save_top_k=1,
    mode="min"
)
```

```python
from lightning.pytorch import Trainer
trainer = Trainer(
    max_epochs=50,            
    accelerator="gpu",
    devices=1,
    gradient_clip_val=0.1,
    enable_progress_bar=False,
    callbacks=[EpochLogger(), checkpoint_callback]
)

trainer.fit(tft, training_dataloader, validation_dataloader)  
```
